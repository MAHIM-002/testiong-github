```python
import matplotlib.pyplot as plt
```

```python
x= df['age']
y= df['oldpeak']
plt.scatter(x,y,c='r')
plt.show
```

```python
x= df['chol']
y= df['thalach']
plt.scatter(x,y,c='r')
plt.show
```

```python
x= df['age']
y= df['thal']
y1= df['chol']
y2= df['oldpeak']
plt.scatter(x,y,c='r')
plt.scatter(y1,y2,c='g')
plt.show
```

```python
x= df['age']
y= df['thal']
y1= df['chol']
y2= df['oldpeak']
plt.scatter(x,y,c='r',label='age & thal')
plt.scatter(x,y2,c='g',label='age & oldpeak ')
plt.scatter(y1,y2,c='b',label='chol & oldpeak')
plt.scatter(y,y2,c='y',label='thal & oldpeak')
plt.legend()
plt.show
```

```python
x= df['age']
y= df['thal']
y1= df['chol']
y2= df['oldpeak']
plt.bar(x,y,color='r',label='age & thal',width=.5)
plt.bar(x,y2,color='g',label='age & oldpeak',width=.5)
plt.scatter(y1,y2,color='b',label='chol & oldpeak',width=.5)
plt.scatter(y,y2,color='y',label='thal & oldpeak',width=.5)
plt.legend()
plt.show
```
